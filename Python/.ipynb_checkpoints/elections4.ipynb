```python
import matplotlib.pyplot as plt
print('\n\t\t------WELCOME TO CHITKARA ELECTIONS------\t\t\n')
A='Candidate A'
B='Candidate B'
C='Candidate C'
D='Candidate D'
print('\n\t\t-----PARTICIPATING CANDIDATES-----\t\t\n')
print('\n\t\t-----',A,'-----\n\t\t-----',B,'-----\n\t\t-----',C,'-----\n\t\t-----',D,'-----')
a=0
b=0
c=0
d=0
total_votes=0
li=[]
ath=[2210991450,2210991437,2210991464]
e=1
n=0
while e>0:
    v=0
    e=int(input("Press 1 to vote\nPress 2 to see results \nPress 3 to finish elections \n\nYOUR CHOICE----->>>   "))
    while v<1:
        if e==1:
            n=int(input("ENTER VOTER ID--->>  "))
            if len(str(n))!=10:
                print("\n\t\t-----INVALID VOTER ID-----")
                break
            else:    
                if n in li:
                    print("\n\t-----ID ALREADY EXISTED-----")
                    break
                else:
                    li.append(n)
                    print("\n\t\t-----WELCOME-----\n\npress 1 for ",A,"\npress 2 for ",B,"\npress 3 for ",C,"\npress 4 for ",D)
                    f=int(input("\nYour response-->"))
                    print('\n\t\t-----Vote Casted-----')
                    if f==1:
                        a=a+1
                        total_votes+=1
                    elif f==2:
                        b=b+1
                        total_votes+=1
                    elif f==3:
                        c=c+1
                        total_votes+=1
                    elif f==4:
                        d=d+1
                        total_votes+=1
                    v+=1
        elif e==2:
            n=int(input("ENTER VOTER ID--->>  "))
            if n in ath:
                print('\n\tVotes of ',A,'are',a)
                print('\tVotes of ',B,'are',b)
                print('\tVotes of ',C,'are',c)
                print('\tVotes of ',D,'are',d)
                print('\tTotal votes casted= ',total_votes)
                if a>b and a>c and a>d:
                    print('\n\t',A," won the elections with",a*100/total_votes,'% votes')
                elif b>a and b>c and b>d:
                    print('\n\t',B,' won the elections with',b*100/total_votes,'% votes')
                elif c>a and c>b and c>d:
                    print('\n\t',C,' won the elections  with',c*100/total_votes,'% votes')
                elif d>a and d>b and d>c:
                    print('\n\t',D,' won the elections with',a*100/total_votes,'% votes')
                elif a==b==c==d:
                    print('\n\tNo candidate won the elections')
                elif a==b==c:
                    print('\n\t',A, B,' and ',C,' tied')
                elif b==c==d:
                    print('\n\t',B, C,' and ',D,' tied')
                elif a==c==d:
                    print('\n\t',A, C,' and ',D,' tied')
                elif a==b==d:
                    print('\n\t',A, B,' and ',D,' tied')
                elif a==b:
                    print('\n\t',A,' and ',B,' tied')
                elif a==c:
                    print('\n\t',A,' and ',C,' tied') 
                elif a==d:
                    print('\n\t',A,' and ',D,' tied') 
                elif b==c:
                    print('\n\t',B,' and ',C,' tied') 
                elif b==d:
                    print('\n\t',B,' and ',D,' tied') 
                elif c==d:
                    print('\n\t',C,' and ',D,' tied')     
                x=['Candidate A','Candidate B','Candidate C','Candidate D']
                h=[a,b,c,d]
                cl=['yellow','brown','red','green']
                plt.pie(h,labels=x,autopct='%2.2f%%',colors=cl)
                plt.show()
                v+=1
            else:
                print('\n\t\t-----Not eligible to see results-----')
                break
                v+=1 
        elif e==3:
            n=int(input("ENTER VOTER ID--->>  "))  
            if n in ath:
                print("\n\t\t-----Elections have ended-----\n\t\t-----Thank You-----")
                e=-1
                break
            else:
                print('\n\t\t-----Not eligible to end elections-----')
                break
        else:
            print('\n\t\t-----Invalid choice-----')
            break
```
